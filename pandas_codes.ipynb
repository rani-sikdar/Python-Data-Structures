movies where audeince score is greater than rotten tomatoes rating

```python
movies[movies['Audience score %'] > movies['Rotten Tomatoes %']].shape[0]
```

```python
x= movies[movies['Audience score %'] > movies['Rotten Tomatoes %']].shape[0]
y = movies.shape[0]

percentage= (x/y)*100
print(round(percentage))
```
