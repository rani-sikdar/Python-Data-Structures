```python
import pandas as pd
```

```python
df = pd.read_csv('/content/CricketTestMatchData.csv')
df.head()
```

Renaming the columns

```python
df = df.rename(columns = {'Mat':'Matches','NO':'Not_Outs', 'HS':'Highest_Inns_Score', 'BF':'Balls_Faced', 'SR':'Batting_Strike_Rate'})
df
```

check null values

```python
df.isnull().any()
```

```python
df[df['Balls_Faced'].isna() == 1]
```

```python
df['Balls_Faced'] = df['Balls_Faced'].fillna(0)
```

```python
df['Batting_Strike_Rate'] = df['Batting_Strike_Rate'].fillna(0)
```

```python
df[df['Player']=='ED Weekes (WI)']
```

check duplicates

```python
df.duplicated()
```

```python
df[df['Player'].duplicated() == 1]
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

split the span column

```python
df['Span'].str.split(pat= '-')
```

```python
df['Span'].str.split(pat= '-').str[0]
```

```python
df['Rookie_Year']= df['Span'].str.split(pat= '-').str[0]
```

```python
df['Final_Year']= df['Span'].str.split(pat= '-').str[1]
```

```python
df
```

```python
df = df.drop(['Span'], axis=1)
```

```python
df.head()
```

Split country from the player

```python
df['Player'].str.split(pat= '(')
```

```python
df['Country'] = df['Player'].str.split(pat= '(').str[1]
```

```python
df.Country
```

```python
df['Country'].str.split(pat= ')').str[0]
```

```python
df['Country'] = df['Country'].str.split(pat= ')').str[0]
```

```python
df.Country
```

```python
# for players

df['Player'].str.split('(')
```

```python
df['Player'] = df['Player'].str.split('(').str[0]
```

```python
df.Player
```

```python
df.head()
```

change datatypes

```python
df.dtypes
```

```python
df['Highest_Inns_Score'].str.split('*').str[0]
```

```python
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].str.split('*').str[0]
```

```python
df['Highest_Inns_Score']
```

```python
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].astype('int')
```

```python
df = df.astype({'Rookie_Year':'int', 'Final_Year':'int'})
```

```python
df['Matches'] = df['Matches'].str.split('*').str[0]
```

```python
df['Matches'] = df['Matches'].astype('int')
```

```python
df.dtypes
```

```python
df['Balls_Faced'] = df['Balls_Faced'].str.split('+').str[0]
```

```python
df.Balls_Faced
```

```python
df['Balls_Faced'] = df['Balls_Faced'].fillna(0)
```

calculate career length

```python
df['Career_Length'] = df['Final_Year'] - df['Rookie_Year']
```

```python
df.head()
```

```python
# calculate average career length

df['Career_Length'].mean()
```

```python
# calculate avg batting strike rate for cricketers who played over 10 years

df[df['Career_Length'] > 10]['Batting_Strike_Rate'].mean()
```

```python
# find number of cricketrs played before 1960

df[df['Rookie_Year'] <1960]['Player'].count()
```

```python
# max highest innings score by country

df.groupby('Country')['Highest_Inns_Score'].max().to_frame('Highinncountry').reset_index().sort_values('Highinncountry', ascending=False)
```

```python
# number of hundreds, fities, ducks avg by country

df.groupby('Country')[['100', '50', '0']].mean()
```
